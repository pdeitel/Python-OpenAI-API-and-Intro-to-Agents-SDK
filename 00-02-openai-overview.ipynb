# OpenAI Python Ecosystem Overview
* **OpenAI Python SDK** for direct API calls from Python programs and notebooks
    * https://github.com/openai/openai-python
    * https://developers.openai.com/api/docs/
* **OpenAI Python Agents SDK** for code-first agent workflows with tools, handoffs, state, and observability
    * https://github.com/openai/openai-agents-python
    * https://developers.openai.com/api/docs/guides/agents/quickstart
* **Codex** as a coding agent for working on real software projects and course code
    * Also, enables capabilities like controlling your local computer
    * https://developers.openai.com/codex
---

## OpenAI Python SDK
* Python client library for calling OpenAI APIs
* You orchestrate directly in your Python apps:
    * Text and code generation with the Responses API
    * Vision and multimodal prompts
    * Structured outputs and function calling
    * Audio, image, moderation, and tool-using workflows
    * ...
* Responses API: Direct Model Workflows
    * Central API  
    * Prompting a model and processing its returned content
    * Asking for JSON or another structured result
    * Giving models tools while your app maintains control flow

---

## OpenAI Python Agents SDK
* For apps that need more orchestration than a single API call
* Agents can
    * plan multi-step control flow
    * call tools
    * collaborate with other agents as tools or via handoffs
    * manage conversation state to move through multi-step tasks
    * work autonomously or with human-in-the-loop
* Uses the Responses API under the hood

---

## Agents SDK Building Blocks
* **Agents**: instructions, model choice, tools, and behavior boundaries
* **Tools**: functions, hosted tools, and external capabilities exposed to the model
* **Handoffs**: moving work from one specialized agent to another
* **Conversation State**: keeping the right context across a multi-step interaction
* **Tracing and evaluation**: inspecting what happened and improving reliability
