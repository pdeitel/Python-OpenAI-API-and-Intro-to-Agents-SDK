# Content Moderation and the OpenAI Moderation API
* Helps developers identify/filter potentially harmful or inappropriate text/image content
* Important for apps with user-generated content and agent guardrails
    * Input moderation 
    * Output review
    * Repeatable safety checks
    * Code can branch, block, warn, escalate, log for subsequent audits
* Checks text/image inputs for harassment, hate speech, violence, self-harm and sexually explicit content and subcategories
    * https://platform.openai.com/docs/guides/moderation#content-classifications
* For each category/subcategory, response contains:
    * `True`/`False` — whether that category was flagged
    * Confidence score `0.0` to `1.0` — higher indicates more offensive/harmful

---

## `check_prompt` and `display_moderation_result` Functions
* [Open `moderation.py`](./source/moderation.py)
* Submit moderation requests via the `OpenAI` object's `moderations.create` method
* `omni-moderation-latest` — default model 
* Moderation API inputs to moderate
    * One string 
    * List of strings
    * List of dictionaries representing text and images to classify
* `response.results` is a list of `Moderation` objects
    * one element if input is one string
* `display_moderation_results` displays a `Moderation` result
* `Moderation` object's `categories` attribute (a `Categories` object) contains
    * Category names
    * `True`/`False` values indicating the flagged categories
* `Moderation` object's `category_scores` attribute (a `CategoriesScores` object) contains 
    * Same category names 
    * Numeric confidence scores from `0.0` to `1.0`
