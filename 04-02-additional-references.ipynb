## Workspace Agents

* No-code AI agents that run inside ChatGPT workspaces
* Designed for organizations to automate recurring business workflows
* Can:
    * Access approved enterprise data and connectors
    * Execute multi-step tasks autonomously
    * Run scheduled workflows
    * Share agents across an organization
* Managed by workspace administrators with centralized permissions and governance
* Intended for business users rather than application developers
* Recommended by OpenAI for natural-language workflow automation instead of code-based agent development

Availability
* ChatGPT Business
* ChatGPT Enterprise
* ChatGPT Edu
* ChatGPT Teachers (where available)
