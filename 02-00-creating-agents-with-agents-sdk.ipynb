# Creating Agents with the OpenAI Agents SDK

* OpenAI Agents SDK is a lightweight package for building agentic AI apps
* Use Python APIs/SDK for simple, direct model calls
* Use Agents SDK to conveniently orchestrate multi-step work 
 
---

## Agents SDK Features 
* **Agent** — instructions + LLM + tools
    * Stateless by default
    * Instructions — role, boundaries, tone, success criteria/outputs
    * LLM/model reasons about task and chooses next actions
* **Tools** — Extend model's capabilities
    * OpenAI hosted tools
    * Custom function tools (any Python function)
    * MCP (model context protocol) tools hosted elsewhere
    * Other agents as tools
* **State** — Lets multi-turn work preserve useful context
* **Handoffs** — Tranfer control to other agents
* **Guardrails** — Input/output validation and safety checks
* **Tracing** — OpenAI Agents auto-log steps 
    * Visualizes workflow
    * Shows decisions, tool calls (not hosted tools)
    * Helps you evaluate, fine-tune and debug agent apps
* **Runner** — The **Agent loop** 
    * Calls agent's LLM model with input
    * Inspects model output
    * If model says to call tools, executes them
    * If model says to hand off to another agent, continues with that agent
    * Auto loops until final answer produced, then returns result

### More Advanced Agents
* **Sandbox Agents** — Work in a controlled filesystem/workspace **\[in beta, may change\]**
* **Realtime Agents** — Voice agents via `gpt-realtime-1.5` model
---

## The ReAct Pattern — **Re**asoning + **Act**ing 
* Foundational loop behind modern AI agents.
* Introduced by Yao et al. (2022)
    * [ReAct: Synergizing Reasoning and Acting in Language Models](https://arxiv.org/abs/2210.03629) 
* Interleaves model reasoning with tool calls
* Feeds results back to model as context for next reasoning step
* Model decides whether to 
    * Returns results 
    * Updates its plan in real time based on what it observes and continue
---

## Sample ReAct Loop
Simple request for latest AAPL stock price
* Thought -> "I need the current price of AAPL to answer this."
* Action  -> call tool `get_quote('AAPL')`
* Observe -> `{'last_price': 211.34, 'currency': 'USD', ...}`
* Thought -> "Now I have the price. I can answer the question."
* Answer  -> "Apple is currently trading at $211.34."

### Notes
* Each observation is part of input to next reasoning step
* If a tool fails or returns unexpected data, model can detect it and try a different approach — no special error-handling code required
* OpenAI doesn't use the term ReAct in their documentation, but their agent loop does this.
