# Agents SDK Examples Study Order
* Source tree: [OpenAI Agents SDK examples](https://github.com/openai/openai-agents-python/tree/main/examples)



## Study Order Notes
* OpenAI does not provide an SDK examples roadmap
* May of the examples have no comments/documentation
* The script OpenAI provides to run the examples, simply goes through them in alphabetical order by folder and file name



## How I Created This Document
* Tools like Codex, Claude Code, ..., can understand and explain code bases
* Asked Codex to study all examples and recommend order based on difficulty and features
* Also asked for a brief description of each

### `examples/basic/`

1. `hello_world.py` (yes)
    * Introduces the smallest runnable Agents SDK flow: define an `Agent`, call `Runner`, and inspect the final answer.
2. `hello_world_jupyter.ipynb` (yes)
    * Notebook version of the first agent run. Readers see how the same `Agent` plus `Runner` pattern works in an interactive teaching environment.
3. `hello_world_gpt_5.py` (yes)
    * Adds explicit GPT-5 model and reasoning settings to the basic agent. Readers see how model choice and `ModelSettings` affect an otherwise simple run.
4. `dynamic_system_prompt.py` (maybe)
    * Shows runtime-generated instructions from context, so one agent can adapt behavior without separate definitions.
5. `tools.py` (yes)
    * Introduces `@function_tool` with a callable the model can invoke. Readers learn the first step from plain text generation toward agents that use application code.
6. `usage_tracking.py` (maybe)
    * Shows where token and request usage are reported after an agent run.
7. `stream_text.py` (yes)
    * Demonstrates streaming text deltas for responsive terminal or notebook output.
8. `stream_items.py` (maybe)
    * Demonstrates higher-level streamed run items such as messages, tool calls, and events.
9. `stream_function_call_args.py` (advanced)
    * Shows how to inspect function-call arguments while the model is still producing them.
10. `previous_response_id.py`
    * Continues a conversation with a prior Responses API response ID instead of resending all history.
11. `local_image.py`
    * Passes a local image as base64 input for multimodal image understanding.
12. `remote_image.py`
    * Passes a hosted image URL as input for a simple vision workflow.
13. `local_file.py`
    * Passes local file content directly into an agent run without a vector store.
14. `remote_pdf.py`
    * Analyzes a PDF by URL, demonstrating direct remote-document input.
15. `image_tool_output.py`
    * Returns image content from a tool, showing that tools can produce multimodal outputs.
16. `non_strict_output_type.py`
    * Explores structured output when a schema cannot use strict JSON mode.
17. `tool_guardrails.py`
    * Combines tool use with guardrail checks so tool workflows are constrained by policy or validation.
18. `agent_lifecycle_example.py`
    * Adds hooks for observing agent/run lifecycle events, tool calls, model responses, and usage.
19. `lifecycle_example.py`
    * Adds hooks for observing agent/run lifecycle events, tool calls, model responses, and usage.
20. `prompt_template.py`
    * Uses a platform prompt template with runtime variables.
21. `retry.py`
    * Configures model retry behavior for transient failures.
22. `retry_litellm.py`
    * Applies retry behavior while using a LiteLLM-backed provider.
23. `hello_world_gpt_oss.py`
    * Runs an agent against a local `gpt-oss` model through an OpenAI-compatible client. Readers learn how the SDK abstractions map to local open-weight model serving.
24. `stream_ws.py`
    * Shows a WebSocket-oriented streaming pattern for service or browser integrations.

### `examples/agent_patterns/`
OpenAI’s README frames this folder around key agent patterns 
* deterministic flows
* routing/handoffs
* agents-as-tools
* judging
* parallelization
* guardrails
* human-in-the-loop patterns

1. `README.md`
    * Provides setup notes and conceptual framing for the examples in that folder before readers inspect the code.
2. `deterministic.py`
    * Breaks work into fixed Python-orchestrated agent steps.
3. `routing.py`
    * Uses a triage/routing agent and handoffs to specialized agents.
4. `agents_as_tools.py`
    * Uses specialized agents as callable tools instead of handing off control.
5. `agents_as_tools_structured.py`
    * Calls specialized agents as tools with structured input contracts.
6. `agents_as_tools_streaming.py`
    * Streams progress from nested agent-as-tool calls.
7. `agents_as_tools_conditional.py`
    * Enables or disables agent tools based on runtime context or access level.
8. `parallelization.py`
    * Runs multiple agents in parallel and selects or synthesizes the best result.
9. `llm_as_a_judge.py`
    * Uses one model call to judge or improve another model call.
10. `input_guardrails.py`
    * Checks user input before or alongside the main agent run.
11. `output_guardrails.py`
    * Validates final output before returning it to the caller.
12. `streaming_guardrails.py`
    * Checks streamed output during generation so bad output can be stopped early.
13. `forcing_tool_use.py`
    * Forces tool use and compares ways to turn tool results into final output.
14. `human_in_the_loop.py`
    * Serializes an interrupted run, collects approval, and resumes execution.
15. `human_in_the_loop_custom_rejection.py`
    * Customizes how rejected tool approvals are reported back into the run.
16. `human_in_the_loop_stream.py`
    * Combines streaming output with approval interruptions.
