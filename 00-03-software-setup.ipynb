## Installing Course Dependencies
For Anaconda users:
* **Anaconda/macOS:** `bash setup/setup_mac.sh`
* **Anaconda/Windows:** `setup\setup_windows.bat`

For other Python distribution users:
* **pip/macOS or Linux:** `bash setup/setup_pip_mac.sh`
* **pip/Windows:** `setup\setup_pip_windows.bat`

---


## What the Setup Installs
* OpenAI Python SDK: `openai`
* OpenAI Agents SDK with LiteLLM support: `openai-agents[litellm]`
* JupyterLab and the `Python (deitel-openai)` kernel
* Graphviz support for Agents SDK workflow visualizations
* Playwright Chromium browser for the Computer Tool demo
* Common data-science, visualization and NLP libraries such as `pandas`, `matplotlib`, `seaborn`, `scikit-learn`, `spacy` and `nltk`
* spaCy small English model: `en_core_web_sm`
